```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw = pd.read_excel("/Users/nataschajademinnitt/Documents/5. Learning/Distribution of Things/distributionofthings.analysis/good_reads/data/goodreadsdata.xlsx")
```

```python
raw = raw[raw['ratings_count'] > 500]
raw = raw[raw['language_code'].isin(['en-CA', 'en-GB', 'en-US', 'eng'])]
raw.rename(columns={'  num_pages' : 'num_pages'}, inplace=True)
books = raw[['title', 'authors', 'average_rating', 'num_pages', 'ratings_count', 'publication_date']].copy()
```

```python
sns.boxplot(data=books, x='average_rating')
plt.show()
```
