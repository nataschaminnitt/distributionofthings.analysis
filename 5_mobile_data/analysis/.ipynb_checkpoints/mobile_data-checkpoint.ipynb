## IV. Data visualisation

### Worldwide data costs (2019 – 2023) 

```python
# Melting the years to compare changes over time
df_melted = df.melt(value_vars=['2023', '2022', '2021', '2020', '2019'],
                    var_name='year', 
                    value_name='value')
```

```python
# Set figure
sns.set_style('white')
sns.set_context('paper')
plt.figure(figsize=(8, 5))

# Create figure
order = ['2019', '2020', '2021', '2022', '2023']
sns.boxplot(data=df_melted, y='year', x='value', hue='year', palette='crest', order=order, showfliers=False)

sns.despine(fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False)

# Label
plt.title('Worldwide Cost of 1 GB of Mobile Data', fontsize=16, y=1.05)
plt.xlabel('Cost in USD', fontsize=10, fontweight='bold')
plt.ylabel('')
plt.xlim(-3, 25)

# Annotate
for year in order:
    year_data = df_melted[df_melted['year'] == year]['value']
    q1 = np.percentile(year_data.dropna(), 25)
    q3 = np.percentile(year_data.dropna(), 75)
    iqr = q3 - q1
    lower_whisker = max(year_data[year_data >= (q1 - 1.5 * iqr)].min(), year_data.min())
    upper_whisker = min(year_data[year_data <= (q3 + 1.5 * iqr)].max(), year_data.max())
    plt.annotate(f'$ {lower_whisker:.2f}', xy=(lower_whisker, year), xytext=(-30, -5),
                 textcoords='offset points', ha='center', color='grey')
    plt.annotate(f'$ {upper_whisker:.2f}', xy=(upper_whisker, year), xytext=(30, -5),
                 textcoords='offset points', ha='center', color='grey')

# Citations
plt.subplots_adjust(bottom=0.3)
plt.figtext(0.3, -0.01, "Visualization by distributionofthings.com using Python (Pandas & Seaborn), data by cable.co.uk", ha="center", fontsize=8, color='grey')


# Show the plot
plt.tight_layout()
plt.savefig(figure_path + "worldwide-cost-1gb.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set figure style
sns.set_style('whitegrid')
sns.set_context('paper')
plt.figure(figsize=(8, 4))

# Define the order of years for the x-axis
order = ['2019', '2020', '2021', '2022', '2023']

# Create the box plot with year on the x-axis
sns.boxplot(data=df_melted, x='year', y='value', hue='year', palette='crest', order=order, showfliers=False, linewidth=0.75)

sns.despine(top=True, right=True, left=False, bottom=False)

# Label
plt.title('Changes in the worldwide cost of 1 GB of mobile data from 2019 to 2023', fontsize=14, y=1.05)
plt.xlabel('')
plt.ylabel('', fontsize=10, fontweight='bold')
plt.ylim(-3, 25)
plt.grid(True, axis='y', linestyle='-', linewidth=0.75)
plt.yticks(np.arange(0, 26, 10), [f"$ {i}" for i in range(0, 26, 10)])

# Annotate whiskers with min and max values for each year
for year in order:
    year_data = df_melted[df_melted['year'] == year]['value']
    q1 = np.percentile(year_data.dropna(), 25)
    q3 = np.percentile(year_data.dropna(), 75)
    iqr = q3 - q1
    lower_whisker = max(year_data[year_data >= (q1 - 1.5 * iqr)].min(), year_data.min())
    upper_whisker = min(year_data[year_data <= (q3 + 1.5 * iqr)].max(), year_data.max())
    plt.annotate(f'$ {lower_whisker:.2f}', xy=(year, lower_whisker), xytext=(0, -12),
                 textcoords='offset points', ha='center', color='black', fontsize=8)
    plt.annotate(f'$ {upper_whisker:.2f}', xy=(year, upper_whisker), xytext=(0, 5),
                 textcoords='offset points', ha='center', color='black', fontsize=8)

# Citations
plt.subplots_adjust(bottom=0.3)
plt.figtext(0.1, -0.02, "Costs are presented in USD", ha="center", fontsize=8, color='grey')
plt.figtext(0.32, -0.05, "Visualization by distributionofthings.com using Python (Pandas & Seaborn), data source cable.co.uk", ha="center", fontsize=8, color='grey')

# Show the plot
plt.tight_layout()
plt.savefig(figure_path + "worldwide-cost-1gb.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df_diff = df.copy()

# Drop rows with NaN values
df_diff = df_diff.dropna(subset=['2023'])

# Calculate the standard deviation and mean values for 2023
total_std = round(df_diff['2023'].std(), 2)
total_mean = round(df_diff['2023'].mean(), 2)

# Create a mean difference column and a standardized difference column
df_diff['mean_diff'] = df_diff['2023'] - total_mean
df_diff['std_diff'] = df_diff['mean_diff'] / total_std

# Display the result
df_diff.head()
```

```python
# Creating a dataframe of selected countries
df_diff = df_diff.sort_values('std_diff', ascending=True)

df_sub = df_diff.iloc[::18] # Select every 18th row from the dataframe
df_min = df_diff.nlargest(1, 'std_diff') # Make sure the country with the highest cost is included
df_selected = pd.concat([df_sub, df_min]) # Concatinate the two dataframes

# Display the result
df_selected.head(12)
```

```python
sns.set_style('whitegrid')
sns.set_context('paper')
plt.figure(figsize=(10, 8))

# Plot
g = sns.catplot(data=df_selected, x='std_diff', y='Name', hue='Name', kind='bar', palette='crest')
g.set_axis_labels("Standard deviations from the global average", "")
plt.grid(False)

# Vertical line
plt.axvline(x=0, color='lightgray', linestyle='-', linewidth=1)

# Horizontal line
burkina_faso_position = df_selected['Name'].tolist().index('Burkina Faso')
plt.axhline(y=burkina_faso_position - 0.5, color='lightgray', linestyle='-', linewidth=1)

# Adjust x-axis limits
plt.xlim([-4, 9.5])
g.legend.remove()

# Labelling
plt.title('Cost of 1 GB of mobile data relative to the global average', fontsize=14, ha='center', y=1.03, x=0.45)

# Annotate
plt.text(2, burkina_faso_position + 0.1, '→ More than $ 2.59 ', ha='center', fontsize=9, color='grey')
plt.text(-2, burkina_faso_position - 0.9, 'Less than $ 2.59 ←', ha='center', fontsize=9, color='grey')

israel_position = df_selected['Name'].tolist().index('Israel')
plt.text(-1.2, israel_position + 0.15, '$ 0.02 ', ha='center', fontsize=9, color='grey')
zimbabwe_position = df_selected['Name'].tolist().index('Zimbabwe')
plt.text(8.8, zimbabwe_position + 0.15, '$ 43.75', ha='center', fontsize=9, color='grey')

# Citation
plt.subplots_adjust(bottom=0.3)
plt.figtext(0.43, -0.02, f"In 2023, the global average for 1 GB of data is USD {total_mean} and 1 standard deviation is USD {total_std}", ha="center", fontsize=8, color='grey')
plt.figtext(0.47, -0.05, "Visualization by distributionofthings.com using Python (Pandas & Seaborn), data source by cable.co.uk", ha="center", fontsize=8, color='grey')

# Save the figure with proper spacing
plt.tight_layout()
plt.savefig(figure_path + "data-2023-top-bottom.png", dpi=300, bbox_inches='tight')
plt.show()
```
